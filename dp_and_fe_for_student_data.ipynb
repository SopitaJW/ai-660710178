**‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏õ‡∏£‡∏∞‡πÄ‡∏†‡∏ó‡∏´‡∏°‡∏ß‡∏î‡∏´‡∏°‡∏π‡πà (‡∏ä‡∏∑‡πà‡∏≠, ‡πÄ‡∏û‡∏®, ‡∏Ñ‡∏ì‡∏∞)**

‚Üí ‡∏Ñ‡∏ß‡∏£‡πÉ‡∏ä‡πâ ‡∏ß‡∏¥‡∏ò‡∏µ‡∏ó‡∏µ‡πà 3 ‡πÄ‡∏ï‡∏¥‡∏°‡∏Ñ‡πà‡∏≤‡∏î‡πâ‡∏ß‡∏¢‡∏Ñ‡πà‡∏≤‡∏ó‡∏µ‡πà‡∏û‡∏ö‡∏ö‡πà‡∏≠‡∏¢‡∏ó‡∏µ‡πà‡∏™‡∏∏‡∏î ‡πÄ‡∏û‡∏£‡∏≤‡∏∞‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏õ‡∏£‡∏∞‡πÄ‡∏†‡∏ó‡∏Ç‡πâ‡∏≠‡∏Ñ‡∏ß‡∏≤‡∏°‡πÑ‡∏°‡πà‡∏°‡∏µ‡∏Ñ‡πà‡∏≤‡πÄ‡∏â‡∏•‡∏µ‡πà‡∏¢

**‡πÅ‡∏ñ‡∏ß‡∏ó‡∏µ‡πà‡∏Ñ‡πà‡∏≤‡∏´‡∏≤‡∏¢‡∏´‡∏•‡∏≤‡∏¢‡∏Ñ‡∏≠‡∏•‡∏±‡∏°‡∏ô‡πå (‡πÄ‡∏ä‡πà‡∏ô ‡πÅ‡∏ñ‡∏ß‡∏ó‡∏µ‡πà ‡∏ä‡∏∑‡πà‡∏≠, ‡πÄ‡∏û‡∏®, ‡∏Ñ‡∏ì‡∏∞, GPA, ‡∏ô‡πâ‡∏≥‡∏´‡∏ô‡∏±‡∏Å ‡∏´‡∏≤‡∏¢‡∏û‡∏£‡πâ‡∏≠‡∏°‡∏Å‡∏±‡∏ô)**

‚Üí ‡∏≠‡∏≤‡∏à‡πÉ‡∏ä‡πâ ‡∏ß‡∏¥‡∏ò‡∏µ‡∏ó‡∏µ‡πà 1 ‡∏•‡∏ö‡πÅ‡∏ñ‡∏ß ‡πÄ‡∏û‡∏£‡∏≤‡∏∞‡πÄ‡∏ï‡∏¥‡∏°‡∏´‡∏•‡∏≤‡∏¢‡∏Ñ‡πà‡∏≤‡∏û‡∏£‡πâ‡∏≠‡∏°‡∏Å‡∏±‡∏ô‡∏≠‡∏≤‡∏à‡∏ó‡∏≥‡πÉ‡∏´‡πâ‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡πÑ‡∏°‡πà‡πÅ‡∏°‡πà‡∏ô

```python
# 1. ‡∏•‡∏ö‡πÅ‡∏ñ‡∏ß‡∏ó‡∏µ‡πà‡∏´‡∏≤‡∏¢‡∏°‡∏≤‡∏Å‡∏Å‡∏ß‡πà‡∏≤ 3 ‡∏Ñ‡∏≠‡∏•‡∏±‡∏°‡∏ô‡πå
df_clean = df[df.isnull().sum(axis=1) <= 3].copy()

# 2. ‡πÄ‡∏ï‡∏¥‡∏°‡∏Ñ‡πà‡∏≤‡∏î‡πâ‡∏ß‡∏¢‡∏Ñ‡πà‡∏≤‡πÄ‡∏â‡∏•‡∏µ‡πà‡∏¢‡∏™‡∏≥‡∏´‡∏£‡∏±‡∏ö‡∏ï‡∏±‡∏ß‡πÄ‡∏•‡∏Ç
num_cols = df_clean.select_dtypes(include=['float64', 'int64']).columns
for col in num_cols:
    df_clean[col] = df_clean[col].fillna(df_clean[col].mean())

# 3. ‡πÄ‡∏ï‡∏¥‡∏°‡∏Ñ‡πà‡∏≤‡∏î‡πâ‡∏ß‡∏¢‡∏Ñ‡πà‡∏≤‡∏ó‡∏µ‡πà‡∏û‡∏ö‡∏ö‡πà‡∏≠‡∏¢‡∏ó‡∏µ‡πà‡∏™‡∏∏‡∏î‡∏™‡∏≥‡∏´‡∏£‡∏±‡∏ö‡∏´‡∏°‡∏ß‡∏î‡∏´‡∏°‡∏π‡πà
cat_cols = df_clean.select_dtypes(include=['object']).columns
for col in cat_cols:
    df_clean[col] = df_clean[col].fillna(df_clean[col].mode()[0])
```

```python
# ‡∏õ‡∏£‡∏±‡∏ö‡∏ó‡∏®‡∏ô‡∏¥‡∏¢‡∏°‡πÄ‡∏õ‡πá‡∏ô 2 ‡∏ï‡∏≥‡πÅ‡∏´‡∏ô‡πà‡∏á
df_clean = df_clean.round(2)

# ‡∏•‡∏ö‡∏ó‡∏®‡∏ô‡∏¥‡∏¢‡∏° (‡∏õ‡∏±‡∏î‡πÄ‡∏õ‡πá‡∏ô‡∏à‡∏≥‡∏ô‡∏ß‡∏ô‡πÄ‡∏ï‡πá‡∏°)
df_clean["‡∏≠‡∏≤‡∏¢‡∏∏"] = df_clean["‡∏≠‡∏≤‡∏¢‡∏∏"].round(0).astype(int)
df_clean["‡∏ä‡∏±‡πà‡∏ß‡πÇ‡∏°‡∏á‡πÄ‡∏£‡∏µ‡∏¢‡∏ô_‡∏ï‡πà‡∏≠‡∏™‡∏±‡∏õ‡∏î‡∏≤‡∏´‡πå"] = df_clean["‡∏ä‡∏±‡πà‡∏ß‡πÇ‡∏°‡∏á‡πÄ‡∏£‡∏µ‡∏¢‡∏ô_‡∏ï‡πà‡∏≠‡∏™‡∏±‡∏õ‡∏î‡∏≤‡∏´‡πå"].round(0).astype(int)
```

```python
def standardize_faculty(name):
    if name in ['‡∏ß‡∏¥‡∏ó‡∏¢‡∏≤‡∏®‡∏≤‡∏™‡∏ï‡∏£‡πå', '‡∏ß‡∏¥‡∏ó‡∏¢‡πå', 'Science', '‡∏Ñ‡∏ì‡∏∞‡∏ß‡∏¥‡∏ó‡∏¢‡∏≤‡∏®‡∏≤‡∏™‡∏ï‡∏£‡πå' , 'Sci']:
        return '‡∏ß‡∏¥‡∏ó‡∏¢‡∏≤‡∏®‡∏≤‡∏™‡∏ï‡∏£‡πå'
    elif name in ['‡∏®‡∏¥‡∏•‡∏õ‡∏®‡∏≤‡∏™‡∏ï‡∏£‡πå', '‡∏®‡∏¥‡∏•‡∏õ‡πå', 'Arts', 'Liberal Arts']:
        return '‡∏®‡∏¥‡∏•‡∏õ‡∏®‡∏≤‡∏™‡∏ï‡∏£‡πå'
    elif name in ['‡∏ß‡∏¥‡∏®‡∏ß‡∏∞', 'Engineering', '‡∏ß‡∏¥‡∏®‡∏ß‡∏Å‡∏£‡∏£‡∏°' , '‡∏ß‡∏¥‡∏®‡∏ß‡∏Ø',]:
        return '‡∏ß‡∏¥‡∏®‡∏ß‡∏Å‡∏£‡∏£‡∏°‡∏®‡∏≤‡∏™‡∏ï‡∏£‡πå'
    else:
        return name

# ‡πÉ‡∏ä‡πâ‡∏á‡∏≤‡∏ô
df['‡∏Ñ‡∏ì‡∏∞_‡∏°‡∏≤‡∏ï‡∏£‡∏ê‡∏≤‡∏ô'] = df['‡∏Ñ‡∏ì‡∏∞'].apply(standardize_faculty)
```

```python
def standardize_gender(gender):
    if pd.isna(gender):
        return gender
    if gender in ['‡∏ä‡∏≤‡∏¢', 'Male', 'M', '‡∏ä']:
        return '‡∏ä‡∏≤‡∏¢'
    elif gender in ['‡∏´‡∏ç‡∏¥‡∏á', 'Female', 'F', '‡∏ç']:
        return '‡∏´‡∏ç‡∏¥‡∏á'
    else:
        return gender

df['‡πÄ‡∏û‡∏®_‡∏°‡∏≤‡∏ï‡∏£‡∏ê‡∏≤‡∏ô'] = df['‡πÄ‡∏û‡∏®'].apply(standardize_gender)
```

```python
print("‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏ï‡πâ‡∏ô‡∏â‡∏ö‡∏±‡∏ö (‡∏™‡∏£‡∏∏‡∏õ)")
print(df["‡∏≠‡∏≤‡∏¢‡∏∏"].describe())

# ‡∏Ñ‡∏≥‡∏ô‡∏ß‡∏ì IQR
Q1 = df["‡∏≠‡∏≤‡∏¢‡∏∏"].quantile(0.25)
Q3 = df["‡∏≠‡∏≤‡∏¢‡∏∏"].quantile(0.75)
IQR = Q3 - Q1

# ‡∏Ç‡∏≠‡∏ö‡πÄ‡∏Ç‡∏ï
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
print(f"\n‡∏Ç‡∏≠‡∏ö‡πÄ‡∏Ç‡∏ï‡∏•‡πà‡∏≤‡∏á: {lower_bound}")
print(f"‡∏Ç‡∏≠‡∏ö‡πÄ‡∏Ç‡∏ï‡∏ö‡∏ô: {upper_bound}")

# ‡∏Å‡∏£‡∏≠‡∏á‡∏Ñ‡πà‡∏≤‡∏ú‡∏¥‡∏î‡∏õ‡∏Å‡∏ï‡∏¥‡∏≠‡∏≠‡∏Å‡πÑ‡∏õ
df_no_outliers = df[(df["‡∏≠‡∏≤‡∏¢‡∏∏"] >= lower_bound) & (df["‡∏≠‡∏≤‡∏¢‡∏∏"] <= upper_bound)]

print("\n‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏´‡∏•‡∏±‡∏á‡∏Å‡∏£‡∏≠‡∏á‡∏≠‡∏≤‡∏¢‡∏∏ (‡∏™‡∏£‡∏∏‡∏õ)")
print(df_no_outliers["‡∏≠‡∏≤‡∏¢‡∏∏"].describe())

df = df_no_outliers
```

```python
# ‡πÅ‡∏Å‡πâ‡πÑ‡∏Ç GPA ‡∏ú‡∏¥‡∏î‡∏õ‡∏Å‡∏ï‡∏¥
df.loc[df['GPA'] < 0, 'GPA'] = np.nan
df.loc[df['GPA'] > 4, 'GPA'] = 4.0
mean_gpa = df['GPA'].mean()

df['GPA'].fillna(mean_gpa, inplace=True)
```

```python
# 1) ‡∏°‡∏≤‡∏ï‡∏£‡∏ê‡∏≤‡∏ô‡∏Ñ‡πà‡∏≤‡∏Ç‡∏≠‡∏á "‡πÄ‡∏û‡∏®"
gender_map = {
    "‡∏ä‡∏≤‡∏¢": "‡∏ä‡∏≤‡∏¢", "Male": "‡∏ä‡∏≤‡∏¢", "M": "‡∏ä‡∏≤‡∏¢", "‡∏ä": "‡∏ä‡∏≤‡∏¢",
    "‡∏´‡∏ç‡∏¥‡∏á": "‡∏´‡∏ç‡∏¥‡∏á", "Female": "‡∏´‡∏ç‡∏¥‡∏á", "F": "‡∏´‡∏ç‡∏¥‡∏á", "‡∏ç": "‡∏´‡∏ç‡∏¥‡∏á"
}
df["‡πÄ‡∏û‡∏®"] = df["‡πÄ‡∏û‡∏®"].replace(gender_map)

# -------------------------
# 2) ‡∏°‡∏≤‡∏ï‡∏£‡∏ê‡∏≤‡∏ô‡∏Ñ‡πà‡∏≤‡∏Ç‡∏≠‡∏á "‡∏Ñ‡∏ì‡∏∞"
faculty_map = {
    "‡∏ß‡∏¥‡∏ó‡∏¢‡∏≤‡∏®‡∏≤‡∏™‡∏ï‡∏£‡πå": "‡∏ß‡∏¥‡∏ó‡∏¢‡∏≤‡∏®‡∏≤‡∏™‡∏ï‡∏£‡πå", "‡∏ß‡∏¥‡∏ó‡∏¢‡πå": "‡∏ß‡∏¥‡∏ó‡∏¢‡∏≤‡∏®‡∏≤‡∏™‡∏ï‡∏£‡πå", "Science": "‡∏ß‡∏¥‡∏ó‡∏¢‡∏≤‡∏®‡∏≤‡∏™‡∏ï‡∏£‡πå", "‡∏Ñ‡∏ì‡∏∞‡∏ß‡∏¥‡∏ó‡∏¢‡∏≤‡∏®‡∏≤‡∏™‡∏ï‡∏£‡πå": "‡∏ß‡∏¥‡∏ó‡∏¢‡∏≤‡∏®‡∏≤‡∏™‡∏ï‡∏£‡πå",
    "‡∏ß‡∏¥‡∏®‡∏ß‡∏∞": "‡∏ß‡∏¥‡∏®‡∏ß‡∏Å‡∏£‡∏£‡∏°‡∏®‡∏≤‡∏™‡∏ï‡∏£‡πå", "Engineering": "‡∏ß‡∏¥‡∏®‡∏ß‡∏Å‡∏£‡∏£‡∏°‡∏®‡∏≤‡∏™‡∏ï‡∏£‡πå", "‡∏ß‡∏¥‡∏®‡∏ß‡∏Å‡∏£‡∏£‡∏°": "‡∏ß‡∏¥‡∏®‡∏ß‡∏Å‡∏£‡∏£‡∏°‡∏®‡∏≤‡∏™‡∏ï‡∏£‡πå", "‡∏ß‡∏¥‡∏®‡∏ß‡∏Ø": "‡∏ß‡∏¥‡∏®‡∏ß‡∏Å‡∏£‡∏£‡∏°‡∏®‡∏≤‡∏™‡∏ï‡∏£‡πå", "‡∏ß‡∏¥‡∏®‡∏ß‡∏Å‡∏£‡∏£‡∏°‡∏®‡∏≤‡∏™‡∏ï‡∏£‡πå": "‡∏ß‡∏¥‡∏®‡∏ß‡∏Å‡∏£‡∏£‡∏°‡∏®‡∏≤‡∏™‡∏ï‡∏£‡πå",
    "‡∏®‡∏¥‡∏•‡∏õ‡∏®‡∏≤‡∏™‡∏ï‡∏£‡πå": "‡∏®‡∏¥‡∏•‡∏õ‡∏®‡∏≤‡∏™‡∏ï‡∏£‡πå", "Arts": "‡∏®‡∏¥‡∏•‡∏õ‡∏®‡∏≤‡∏™‡∏ï‡∏£‡πå", "‡∏®‡∏¥‡∏•‡∏õ‡πå": "‡∏®‡∏¥‡∏•‡∏õ‡∏®‡∏≤‡∏™‡∏ï‡∏£‡πå", "Liberal Arts": "‡∏®‡∏¥‡∏•‡∏õ‡∏®‡∏≤‡∏™‡∏ï‡∏£‡πå", "Sci": "‡∏ß‡∏¥‡∏ó‡∏¢‡∏≤‡∏®‡∏≤‡∏™‡∏ï‡∏£‡πå"
}
df["‡∏Ñ‡∏ì‡∏∞"] = df["‡∏Ñ‡∏ì‡∏∞"].replace(faculty_map)

# -------------------------
# 3) One-Hot Encoding
df_encoded = pd.get_dummies(df, columns=["‡πÄ‡∏û‡∏®", "‡∏Ñ‡∏ì‡∏∞"], prefix=["‡πÄ‡∏û‡∏®", "‡∏Ñ‡∏ì‡∏∞"])
```
