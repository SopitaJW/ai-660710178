**ข้อมูลประเภทหมวดหมู่ (ชื่อ, เพศ, คณะ)**

→ ควรใช้ วิธีที่ 3 เติมค่าด้วยค่าที่พบบ่อยที่สุด เพราะข้อมูลประเภทข้อความไม่มีค่าเฉลี่ย

**แถวที่ค่าหายหลายคอลัมน์ (เช่น แถวที่ ชื่อ, เพศ, คณะ, GPA, น้ำหนัก หายพร้อมกัน)**

→ อาจใช้ วิธีที่ 1 ลบแถว เพราะเติมหลายค่าพร้อมกันอาจทำให้ข้อมูลไม่แม่น

```python
# 1. ลบแถวที่หายมากกว่า 3 คอลัมน์
df_clean = df[df.isnull().sum(axis=1) <= 3].copy()

# 2. เติมค่าด้วยค่าเฉลี่ยสำหรับตัวเลข
num_cols = df_clean.select_dtypes(include=['float64', 'int64']).columns
for col in num_cols:
    df_clean[col] = df_clean[col].fillna(df_clean[col].mean())

# 3. เติมค่าด้วยค่าที่พบบ่อยที่สุดสำหรับหมวดหมู่
cat_cols = df_clean.select_dtypes(include=['object']).columns
for col in cat_cols:
    df_clean[col] = df_clean[col].fillna(df_clean[col].mode()[0])
```

```python
# ปรับทศนิยมเป็น 2 ตำแหน่ง
df_clean = df_clean.round(2)

# ลบทศนิยม (ปัดเป็นจำนวนเต็ม)
df_clean["อายุ"] = df_clean["อายุ"].round(0).astype(int)
df_clean["ชั่วโมงเรียน_ต่อสัปดาห์"] = df_clean["ชั่วโมงเรียน_ต่อสัปดาห์"].round(0).astype(int)
```

# 2. การทำให้ข้อมูลเป็นมาตรฐาน

เหตุผลที่ต้องทำขั้นตอนนี้

คือเพื่อให้ข้อมูลที่มีความหมายเหมือนกันถูกเขียนในรูปแบบเดียวกันทั้งหมด เพื่อป้องกันปัญหาที่เกิดจากความไม่สม่ำเสมอ
และ ขั้นตอนนี้ทำให้ข้อมูล สะอาด + สม่ำเสมอ + พร้อมใช้งาน โดยไม่เกิดความเข้าใจผิดจากความแตกต่างในรูปแบบการเขียน

⬇️โค้ดนี้มีไว้ทำให้ชื่อคณะเขียนเหมือนกันหมด เช่น "วิทย์", "Science" หรือ "คณะวิทยาศาสตร์" ก็เปลี่ยนให้เป็น "วิทยาศาสตร์" รวมถึงเพศก็เช่นกัน จากนั้นใช้ .apply() ให้ฟังก์ชันนี้ทำงานกับทุกแถวในคอลัมน์ คณะ แล้วเก็บผลไว้ในคอลัมน์ใหม่ชื่อ คณะ_มาตรฐาน เพื่อให้ข้อมูลคณะเป็นรูปแบบเดียวกันและนำไปวิเคราะห์ต่อได้ง่าย

```python
def standardize_faculty(name):
    if name in ['วิทยาศาสตร์', 'วิทย์', 'Science', 'คณะวิทยาศาสตร์' , 'Sci']:
        return 'วิทยาศาสตร์'
    elif name in ['ศิลปศาสตร์', 'ศิลป์', 'Arts', 'Liberal Arts']:
        return 'ศิลปศาสตร์'
    elif name in ['วิศวะ', 'Engineering', 'วิศวกรรม' , 'วิศวฯ',]:
        return 'วิศวกรรมศาสตร์'
    else:
        return name

# ใช้งาน
df['คณะ_มาตรฐาน'] = df['คณะ'].apply(standardize_faculty)
```

```python
def standardize_gender(gender):
    if pd.isna(gender):
        return gender
    if gender in ['ชาย', 'Male', 'M', 'ช']:
        return 'ชาย'
    elif gender in ['หญิง', 'Female', 'F', 'ญ']:
        return 'หญิง'
    else:
        return gender

df['เพศ_มาตรฐาน'] = df['เพศ'].apply(standardize_gender)
```

# 3. การจัดการค่าผิดปกติ

ต้องจัดการค่าผิดปกติ (Outliers) เพราะช่วยให้ข้อมูลสะท้อนความจริงมากขึ้น ค่าหลุดสูงหรือต่ำเกินไปจะทำให้ค่าเฉลี่ยและการกระจายข้อมูลเพี้ยน ส่งผลให้การวิเคราะห์และโมเดลทำนายผิด ความสัมพันธ์ระหว่างตัวแปรคลาดเคลื่อน อีกทั้งหลายกรณีเกิดจากการกรอกข้อมูลผิดหรือระบบบันทึกผิด การจัดการจึงช่วยเพิ่มความน่าเชื่อถือ ลดผลกระทบต่อโมเดล และทำให้การแสดงผลกราฟหรือสถิติมีสเกลที่อ่านง่าย เห็นรูปแบบข้อมูลปกติชัดเจนขึ้น

```python
print("ข้อมูลต้นฉบับ (สรุป)")
print(df["อายุ"].describe())

# คำนวณ IQR
Q1 = df["อายุ"].quantile(0.25)
Q3 = df["อายุ"].quantile(0.75)
IQR = Q3 - Q1

# ขอบเขต
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
print(f"\nขอบเขตล่าง: {lower_bound}")
print(f"ขอบเขตบน: {upper_bound}")

# กรองค่าผิดปกติออกไป
df_no_outliers = df[(df["อายุ"] >= lower_bound) & (df["อายุ"] <= upper_bound)]

print("\nข้อมูลหลังกรองอายุ (สรุป)")
print(df_no_outliers["อายุ"].describe())

df = df_no_outliers
```

```python
# แก้ไข GPA ผิดปกติ
df.loc[df['GPA'] < 0, 'GPA'] = np.nan
df.loc[df['GPA'] > 4, 'GPA'] = 4.0
mean_gpa = df['GPA'].mean()

df['GPA'].fillna(mean_gpa, inplace=True)
```

# Feature Engineering - การสร้างคุณลักษณะใหม่

# 1. One-Hot Encoding **แปลงข้อมูลประเภทเป็นตัวเลข** **bold text**

**1) มาตรฐานค่า “เพศ”**
เพราะข้อมูลเพศมีหลายแบบที่ความหมายเหมือนกัน เช่น `ชาย, Male, M, ช` หมายถึง “ชาย” เหมือนกัน เราเลยใช้ `gender_map` กับ `.replace()` เพื่อให้เป็นคำเดียวกัน จะได้ไม่งงตอนวิเคราะห์

**2) มาตรฐานค่า “คณะ”**
ชื่อคณะก็มีหลายแบบ เช่น `วิทยาศาสตร์, วิทย์, Science, คณะวิทยาศาสตร์` หมายถึงคณะเดียวกัน ก็ใช้ `faculty_map` รวมให้เหลือรูปแบบเดียว เพื่อป้องกันคอลัมน์ซ้ำ ๆ ตอนทำต่อ

**3) One-Hot Encoding**
พอค่ามาตรฐานแล้ว เราใช้ `pd.get_dummies()` แปลงข้อมูลจากข้อความเป็นตัวเลข 0/1 เช่น คอลัมน์ `เพศ` จะกลายเป็น `เพศ_ชาย` กับ `เพศ_หญิง` ถ้าเพศตรงกับชื่อคอลัมน์ก็จะเป็น 1 ถ้าไม่ตรงก็เป็น 0 ทำให้เอาไปใช้กับโมเดลหรือคำนวณได้ง่าย

```python
# 1) มาตรฐานค่าของ "เพศ"
gender_map = {
    "ชาย": "ชาย", "Male": "ชาย", "M": "ชาย", "ช": "ชาย",
    "หญิง": "หญิง", "Female": "หญิง", "F": "หญิง", "ญ": "หญิง"
}
df["เพศ"] = df["เพศ"].replace(gender_map)

# -------------------------
# 2) มาตรฐานค่าของ "คณะ"
faculty_map = {
    "วิทยาศาสตร์": "วิทยาศาสตร์", "วิทย์": "วิทยาศาสตร์", "Science": "วิทยาศาสตร์", "คณะวิทยาศาสตร์": "วิทยาศาสตร์",
    "วิศวะ": "วิศวกรรมศาสตร์", "Engineering": "วิศวกรรมศาสตร์", "วิศวกรรม": "วิศวกรรมศาสตร์", "วิศวฯ": "วิศวกรรมศาสตร์", "วิศวกรรมศาสตร์": "วิศวกรรมศาสตร์",
    "ศิลปศาสตร์": "ศิลปศาสตร์", "Arts": "ศิลปศาสตร์", "ศิลป์": "ศิลปศาสตร์", "Liberal Arts": "ศิลปศาสตร์", "Sci": "วิทยาศาสตร์"
}
df["คณะ"] = df["คณะ"].replace(faculty_map)

# -------------------------
# 3) One-Hot Encoding
df_encoded = pd.get_dummies(df, columns=["เพศ", "คณะ"], prefix=["เพศ", "คณะ"])
```

# 2. การสร้าง Feature จากวันเวลา

โค้ดนี้จะแปลงคอลัมน์ `วันที่ลงทะเบียน` ให้เป็นชนิดข้อมูลวันเวลา (`datetime64`) เพื่อให้ดึงส่วนประกอบของวันเวลาได้ง่าย จากนั้นใช้ `.dt.dayofweek` เพื่อเอาค่าตัวเลข 0–6 (0=จันทร์, 6=อาทิตย์) แล้ว `.map()` ให้กลายเป็นชื่อวันภาษาไทยตามลำดับที่กำหนดใน `thai_days` ผลลัพธ์คือได้คอลัมน์ใหม่ชื่อ **วันในสัปดาห์\_ลงทะเบียน** ที่บอกว่านักศึกษาลงทะเบียนวันอะไร


```python
df["วันที่ลงทะเบียน"] = pd.to_datetime(df["วันที่ลงทะเบียน"], errors="coerce")


thai_days = ['จันทร์','อังคาร','พุธ','พฤหัสบดี','ศุกร์','เสาร์','อาทิตย์']
df["วันในสัปดาห์_ลงทะเบียน"] = df["วันที่ลงทะเบียน"].dt.dayofweek.map(dict(enumerate(thai_days)))
```

โค้ดนี้จะแปลงคอลัมน์ `วันที่ลงทะเบียน` ให้เป็นชนิดวันเวลา จากนั้นดึงหมายเลขวันในสัปดาห์ออกมา แล้วแมปเป็นชื่อวันภาษาไทยตามที่กำหนด ผลลัพธ์คือได้คอลัมน์ใหม่ `วันในสัปดาห์_ลงทะเบียน` ที่บอกว่านักศึกษาลงทะเบียนในวันอะไร เช่น จันทร์ ศุกร์ หรือเสาร์ ทำให้สามารถนำไปวิเคราะห์พฤติกรรมการลงทะเบียนตามวันในสัปดาห์ได้ง่ายมากขึ้น


# 3. การรวม Features

โค้ดนี้ทำเพราะต้องการสร้างฟีเจอร์ BMI และจัดกลุ่มสุขภาพจากข้อมูลส่วนสูงและน้ำหนัก เพื่อใช้วิเคราะห์หรือเปรียบเทียบสุขภาพของนักศึกษาได้ง่ายขึ้น

สิ่งที่ทำคือ

1.แปลงส่วนสูงจากเซนติเมตรเป็นเมตร
และกันค่าศูนย์เพื่อเลี่ยงหารด้วยศูนย์

2.คำนวณค่า BMI ตามสูตร น้ำหนัก ÷ (ส่วนสูง²) และปัดเป็นทศนิยม 2 ตำแหน่ง

3.แบ่งค่า BMI ออกเป็นช่วง พร้อมใส่ป้ายกำกับ เช่น ผอม, ปกติ, อ้วน, อ้วนมาก

```python
# ป้องกันหารด้วยศูนย์/ค่าว่าง
h_m = (df["ส่วนสูง_cm"] / 100).replace(0, np.nan)

df["BMI"] = df["น้ำหนัก_kg"] / (h_m ** 2)
df["BMI"] = df["BMI"].round(2)

df["สุขภาพ"] = pd.cut(
    df["BMI"],
    bins=[0, 18.5, 25, 30, np.inf],
    labels=['ผอม', 'ปกติ', 'อ้วน', 'อ้วนมาก'],
    right=True, include_lowest=True
)
```

คือได้ 2 คอลัมน์ใหม่ — BMI เป็นตัวเลข และ สุขภาพ เป็นหมวดหมู่ที่บอกสถานะสุขภาพของแต่ละคน
